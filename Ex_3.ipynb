```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(columns=['Artist Name','Track Name'],inplace=True)
```

```python
df['key'].fillna(value=df['key'].mean(),inplace=True)
```

```python
df.drop(columns=['Popularity','acousticness','instrumentalness'],inplace=True)
```

```python
df
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
y
```

```python
from tensorflow import keras
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.models import Sequential
import tensorflow as tf
```

```python
y=keras.utils.to_categorical(y,11)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
x_train.shape
```

```python
model1=Sequential()
model1.add(Dense(100,activation='softmax',input_shape=(11,)))
model1.add(Dense(70,activation='relu'))
model1.add(Dense(50,activation='relu'))
model1.add(Dense(11,activation='softmax'))
```

```python
model1.summary()
```

```python
model1.compile(optimizer='adam',
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model1.fit(x_train,y_train,batch_size=50,epochs=100)
```

```python
score=model1.evaluate(x_test,y_test)
```

```python
model2=Sequential()
model2.add(Dense(100,activation='softmax',input_shape=(11,)))
model2.add(Dense(11,activation='softmax'))
```

```python
model2.compile(optimizer='adam',
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
model2.fit(x_train,y_train,batch_size=500,epochs=100)
```

```python
score2=model2.evaluate(x_test,y_test)
```

```python
model3=Sequential()
model3.add(Dense(100,activation='softmax',input_shape=(11,)))
model3.add(Dense(70,activation='relu'))
model3.add(Dense(70,activation='relu'))
model3.add(Dense(70,activation='relu'))
model3.add(Dense(70,activation='relu'))
model3.add(Dropout(0.5))
model3.add(Dense(70,activation='relu'))
model3.add(Dense(11,activation='softmax'))
```

```python
model3.compile(optimizer='adam',
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
model3.fit(x_train,y_train,batch_size=500,epochs=100)
```

```python
score3=model3.evaluate(x_test,y_test)
```

The dataset I have used is a music genre classifier.
I have built 3 different models by altering the hidden layers and activation functions.

The accuracy is around 46-50%. There is no drastic change in it.
