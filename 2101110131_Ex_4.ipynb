```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(columns=['Artist Name','Track Name'],inplace=True)
```

```python
df['key'].fillna(value=df['key'].mean(),inplace=True)
```

```python
df.drop(columns=['Popularity','acousticness','instrumentalness'],inplace=True)
```

```python
df
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
y
```

```python
from tensorflow import keras
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.models import Sequential
import tensorflow as tf
from tensorflow.keras import regularizers
```

```python
y=keras.utils.to_categorical(y,11)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
x_train.shape
```

```python
model1=Sequential()
model1.add(Dense(100,activation='softmax',input_shape=(11,)))
model1.add(Dense(70,activation='relu'))
model1.add(Dense(50,activation='relu'))
model1.add(Dense(11,activation='softmax'))
```

```python
model1.summary()
```

```python
model1.compile(optimizer='adam',
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model1.fit(x_train,y_train,batch_size=50,epochs=100)
```

```python
score=model1.evaluate(x_test,y_test)
```

```python
model1=Sequential()
model1.add(Dense(100,activation='softmax',input_shape=(11,)))
model1.add(Dense(70,activation='relu',kernel_regularizer=regularizers.L1(0.01)))
model1.add(Dense(50,activation='relu'))
model1.add(Dense(11,activation='softmax'))

model1.compile(optimizer='adam',
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
score=model1.evaluate(x_test,y_test)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop=EarlyStopping(monitor='loss',patience=12,mode="auto")
```

```python
from tensorflow.keras.layers import BatchNormalization
model1=Sequential()
model1.add(Dense(100,activation='softmax',input_shape=(11,)))
model1.add(Dense(70,activation='relu'))
model1.add(Dense(50,activation='relu'))
model1.add(BatchNormalization(axis=-1,
    momentum=0.99,
    epsilon=0.001))
model1.add(Dense(11,activation='softmax'))

model1.compile(optimizer='adam',
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
model1.fit(x_train,y_train,batch_size=50,epochs=100,callbacks=early_stop)
score=model1.evaluate(x_test,y_test)
```

```python
from tensorflow.keras.constraints import MaxNorm
model1=Sequential()
model1.add(Dense(100,activation='softmax',input_shape=(11,)))
model1.add(Dense(70,activation='relu'))
model1.add(Dense(50,activation='relu',kernel_constraint=MaxNorm(max_value=5,axis=0)))
model1.add(Dense(11,activation='softmax'))

model1.compile(optimizer='adam',
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
model1.fit(x_train,y_train,batch_size=50,epochs=100,callbacks=early_stop)
score=model1.evaluate(x_test,y_test)
```
